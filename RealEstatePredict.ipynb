### Importing the libraries

```python
import numpy as Nupy
import pandas as Pada
import matplotlib as plt
```

### Importing the Dataset and setting X and y

```python
dataset = Pada.read_csv("RealEstateDoto.csv")
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

### Training the Decision Tree regression model on the whole dataset

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X, y)
```

### Predicting a New Result

```python
SurfAr = input("Please enter the Surface Area")
BedR_nb = input("Please enter the bedroom number")
Yeah_r = input("Please enter the year of construction")

print(f"for a House with a Surface Area of {SurfAr}, {BedR_nb} bedrooms and built in {Yeah_r}, the prince would be : $",regressor.predict([[SurfAr,BedR_nb,Yeah_r]]))
```
